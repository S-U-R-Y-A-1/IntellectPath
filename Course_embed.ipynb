```python
import pandas as pd
from flashtext import KeywordProcessor
import nltk
import string

# Download the stopwords dataset
nltk.download('stopwords')

# Assuming you have already loaded the DataFrame
df = pd.read_csv('top_10_courses_combined_desc.csv')

# Fetch unique subjects and levels for user input
unique_subjects = df['subject'].unique()
unique_levels = df['Level'].unique()

# Display available subjects for user input
print("Available Subjects:")
for idx, subject in enumerate(unique_subjects, start=1):
    print(f"{idx}. {subject}")

# Take user input for the subject
subject_idx = int(input("\nEnter the number corresponding to the subject: ")) - 1
selected_subject = unique_subjects[subject_idx]

# Display available levels for user input
print("\nAvailable Levels:")
for idx, level in enumerate(unique_levels, start=1):
    print(f"{idx}. {level}")

# Take user input for the level
level_idx = int(input("\nEnter the number corresponding to the level: ")) - 1
selected_level = unique_levels[level_idx]

# Filter the dataset based on user input
filtered_df = df[(df['subject'] == selected_subject) & (df['Level'] == selected_level)]

# Display available courses for user input
print("\nAvailable Courses:")
for idx, (title, course_desc) in enumerate(filtered_df[['title', 'combined_desc']].values, start=1):
    print(f"{idx}. {title} - {course_desc}")

# Take user input for the courses (more than 1, less than 3)
selected_courses_indices = []
while True:
    num_courses = int(input("\nEnter the number of courses (more than 1, less than 3): "))
    if 1 < num_courses < 3:
        for i in range(num_courses):
            course_idx = int(input(f"Enter the number corresponding to course #{i+1}: ")) - 1
            selected_courses_indices.append(course_idx)
        break
    else:
        print("Please enter a valid number of courses.")

# Extract keywords from the selected courses
keyword_processor = KeywordProcessor()
selected_courses = filtered_df.iloc[selected_courses_indices]
for course_desc in selected_courses['combined_desc']:
    # Add stopwords to be excluded
    stop_words = set(stopwords.words('english'))

    # Tokenize and remove punctuation
    tokens = [word.strip(string.punctuation) for word in course_desc.split()]

    # Remove stopwords
    tokens = [word.lower() for word in tokens if word.lower() not in stop_words]

    keyword_processor.add_keywords_from_list(tokens)

# Print the selected courses with their corresponding keywords
print("\nSelected Courses:")
for title, course_desc in selected_courses[['title', 'combined_desc']].values:
    print(f"{title} - {course_desc}")

# Print the extracted keywords for each selected course
print("\nExtracted Keywords:")
for course_desc in selected_courses['combined_desc']:
    extracted_keywords = keyword_processor.extract_keywords(course_desc)
    print(f"Course Keywords: {extracted_keywords}")
```

```python
import nltk
import string
import re

nltk.download('wordnet')

# Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')

# Defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output = [i for i in text if i not in stopwords]
    return output

# Defining the function to remove punctuation
def remove_punctuation(text):
    # Check if the input is not NaN
    if isinstance(text, str):
        punctuation_free = "".join([i for i in text if i not in string.punctuation])
        return punctuation_free
    else:
        return ""

from nltk.stem.porter import PorterStemmer

# Defining the object for stemming
porter_stemmer = PorterStemmer()

# Defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text

# Storing the punctuation-free text
def tokenization(text):
    tokens = re.split('\W+', text)
    return tokens

# Applying function to the column
from nltk.stem import WordNetLemmatizer

# Defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()

# Defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text

# Apply the functions to the 'combined_desc' column, handling NaN values
data = computer_science_df['combined_desc'].apply(lambda x: remove_punctuation(x) if pd.notna(x) else "")
data = data.apply(lambda x: x.lower())
data = data.apply(lambda x: tokenization(x))
data = data.apply(lambda x: remove_stopwords(x))
data = data.apply(lambda x: lemmatizer(x))

data = list(data)
print(data[0])
```
